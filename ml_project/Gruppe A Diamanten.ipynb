```python

#The first column seems to be just index
dataset = dataset.drop(["Unnamed: 0"], axis=1)
dataset.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(dataset[['carat', 'depth', 'table', 'x', 'y', 'z', 'price']])
```

```python
dataset.drop(dataset[dataset["z"]==0].index, inplace=True)
dataset.drop(dataset[dataset["y"]==0].index, inplace=True)
dataset.drop(dataset[dataset["x"]==0].index, inplace=True)
dataset.drop(dataset[dataset["z"]>20].index, inplace=True)
dataset.drop(dataset[dataset["y"]>20].index, inplace=True)
dataset.drop(dataset[dataset["x"]>20].index, inplace=True)
dataset.describe()

```

```python
plt.show
```

```python
sns.pairplot(dataset[['carat', 'depth', 'table', 'x', 'y', 'z', 'price']])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(dataset.corr(numeric_only=True), annot=True, cmap='coolwarm', center=0)
plt.title('Price Correlations')
plt.show()
```
